```
import cv2
img = cv2.imread("imori.jpg")
b = img[:, :, 0].copy()
g = img[:, :, 1].copy()
r = img[:, :, 2].copy()

img[:, :, 0]  = r
img[:, :, 1] = g
img[:, :, 2] = b

cv2.imwrite("out1.jpg", img)
cv2_imshow(img)
```

#2

```
import cv2
img = cv2.imread("imori.jpg").astype(np.float32)
b = img[:, :, 0].copy()
g = img[:, :, 1].copy()
r = img[:, :, 2].copy()

y = 0.2126*r + 0.7152*g + 0.0722*b
out = y.astype(np.uint8)
cv2.imwrite("out2.jpg", out)
cv2_imshow(out)
```

#3

```
import cv2
img = cv2.imread("imori.jpg").astype(np.float32)
b = img[:, :, 0].copy()
g = img[:, :, 1].copy()
r = img[:, :, 2].copy()

y = 0.2126*r + 0.7152*g + 0.0722*b
out = y.astype(np.uint8)

#二値化
th = 128
out[out < th] = 0
out[out >= th] = 255

cv2.imwrite("out3.jpg", out)
cv2_imshow(out)
```
